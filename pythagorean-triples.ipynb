```python
import numpy as np
import matplotlib.pyplot as plt

numberrange = np.arange(2,51)

a = []
b = []
c = []

for m in numberrange:
    for n in range(numberrange[0], m):
        a.append( m**2 - n**2 )
        b.append( 2*m*n )
        c.append( m**2 + n**2 )
```

```python
a
```

```python
import numpy as np

pytcheck = np.array(a)**2 + np.array(b)**2 - np.array(c)**2

plt.plot(pytcheck, 'o')
plt.show()
```

```python
fig,ax = plt.subplots(1,figsize=(10,7))

plt.scatter(a,b,c=c,marker='o',alpha=.9,linewidth=.5,edgecolors='k')
plt.axis('on')
plt.show()
```
